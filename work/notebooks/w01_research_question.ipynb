# 2. The Question

## Research Question

Which pages are visible and not capturing clicks or engagement or should be reviewed as such for metadata, content or monitoring?

## Unit of Analysis

The unit of analysis is **one content page**.

## Decision

The next step is to determine which pages need to be optimized for CTR or engagement.

## Output

The results will be a list of pages that have opportunities for improvement in CTR/Engagement.

## Action

Content teams will be able to analyze page titles, meta descriptions, page structure, content quality, and user experience to boost click-through rates and engagement.

## Cost of a Wrong Recommendation

If the recommendation is wrong, high-performing pages may receive unnecessary changes while pages with real problems could be ignored. This may waste time and reduce the effectiveness of optimization efforts.

## Why Data or Machine Learning Can Help

CTR and engagement are affected by multiple factors such as impressions, average position, page age, content type, sessions, and engagement metrics. Machine learning can analyze these factors together and help prioritize pages more effectively than simple manual rules.
